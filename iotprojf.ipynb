```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("pdata.csv")
df
```

```python
print(df.isnull().sum())
#linear y=mx+b Outcome=m*Insulin +b
plt.scatter(df.Insulin,df.Outcome,color='red',marker='*')
#linear regression training
```

```python
reg_analysis = linear_model.LinearRegression()
reg_analysis.fit(df[['Insulin']],df.Outcome)
```

```python
reg_analysis.predict([[300]])
```

```python
#multiple Outcome=m1*Pregnancies +m2*Glucose +m3*BloodPressure +m4*SkinThickness +m*Insulin+m5*BMI +m6*DiabetesPedigreeFunction +m7*Age +b
reg_analysis1 = linear_model.LinearRegression()
reg_analysis1.fit(df[['Pregnencies','Glucose','BloodPressure','skin thickness','Insulin','BMI','Diabetes','Age']],df.Outcome)
```
